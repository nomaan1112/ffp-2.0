```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\\Projects\\Flight-Fares-Prediction-2.0\\Data\\Updated_dataset.csv')
```

```python
df['airline'].value_counts()
```

```python
airline_counts = df['airline'].value_counts()

plt.figure(figsize=(15, 10))
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%')
plt.title('Airline Market Share')
plt.show()
```

```python
sns.barplot(x='class_type', y='price', data=df)
plt.show()
```

```python
sns.boxplot(x = 'airline', y = 'price', data = df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='class_type',y='price',data=df,palette='hls')
plt.title('Class Vs Ticket Price',fontsize=15)
plt.xlabel('Class',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
average_price_by_class = df.groupby('class_type')['price'].mean()

average_price_by_class.plot(kind='bar')
plt.title('Average Price by Class')
plt.xlabel('Class')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(24,10))
plt.subplot(1,2,1)
sns.boxplot(x='departure_time',y='price',data=df)
plt.title('Departure Time Vs Ticket Price',fontsize=20)
plt.xlabel('Departure Time',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.subplot(1,2,2)
sns.boxplot(x='arrival_time',y='price',data=df,palette='hls')
plt.title('Arrival Time Vs Ticket Price',fontsize=20)
plt.xlabel('Arrival Time',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(24,10))
plt.subplot(1,2,1)
sns.boxplot(x='source_city',y='price',data=df)
plt.title('Source City Vs Ticket Price',fontsize=20)
plt.xlabel('Source City',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.subplot(1,2,2)
sns.boxplot(x='destination_city',y='price',data=df,palette='hls')
plt.title('Destination City Vs Ticket Price',fontsize=20)
plt.xlabel('Destination City',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df,x='duration',y='price',hue='class_type',palette='hls')
plt.title('Ticket Price Versus Flight Duration Based on Class',fontsize=20)
plt.xlabel('Duration',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df,x='days_left',y='price',color='green')
plt.title('Days Left For Departure Versus Ticket Price',fontsize=20)
plt.xlabel('Days Left for Departure',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df,x='days_left',y='price',color='red',hue='airline',palette='hls')
plt.title('Days Left For Departure Versus Ticket Price of each Airline',fontsize=15)
plt.legend(fontsize=12)
plt.xlabel('Days Left for Departure',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```
